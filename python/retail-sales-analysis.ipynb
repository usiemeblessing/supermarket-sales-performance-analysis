### Question 13. Convert order_date to datetime and extract year, month, quarter, day of week

#### DATETIME TRANSFORMATION AND EXTRACTION

```python
# convert order_date to datetime
```

```python
df["order_date"] = pd.to_datetime(df["order_date"])
```

```python
# extract year
df["year"] = df["order_date"].dt.year
```

```python
# extract month
df["month"] = df["order_date"].dt.month_name()
```

```python
# extract quarter
df["quarter"] = df["order_date"].dt.quarter
```

```python
# extract day of week
df["day_of_week"] = df["order_date"].dt.day_name()
```

### Question 14. Calculate a new column: gross_revenue = quantity × unit_price; compare to net_revenue

#### FEATURE ENGINEERING

```python
# create gross revenue
df["gross_revenue"] = df["quantity"] * df["unit_price"]
```

```python
# comparing gross_revenue against net_revenue
df["comparison"] = df["gross_revenue"] == df["net_revenue"]
```

```python
# compare gross_revenue to net_revenue
df[["gross_revenue", "net_revenue", "comparison"]].head()
```

### Question 15. Plot monthly revenue trends as a line chart across all three years

#### VISUALIZATION

##### MONTHLY REVENUE AGRREGATION AND TREND ANALYSIS

```python
# 1. Filter out zero/incomplete years (keep strictly 2021 to 2023)
df_filtered = df[df['year'].isin([2021, 2022, 2023])]

# 2. Run aggregation on filtered data
monthly_revenue = df_filtered.groupby(["year", "month"], observed=False)["gross_revenue"].sum()
monthly_revenue_df = monthly_revenue.unstack(level=0) / 1e8

# 3. Plot the line chart
ax = monthly_revenue_df.plot(kind="line", figsize=(12, 6), marker="o", title="Monthly Revenue Trend (2021 - 2023)")

# 4. Show Y-Axis
ax.yaxis.set_visible(True)
ax.set_ylabel("Revenue (in Billions)", fontsize=10, labelpad=10)
ax.grid(axis='y', linestyle='--', alpha=0.5) # Soft horizontal grid lines

# 5. X-Axis Styling
ax.set_xticks(range(len(monthly_revenue_df)))
ax.set_xticklabels(monthly_revenue_df.index, rotation=45)
ax.set_xlabel("Month", labelpad=10)

# 6. Remove outer borders
for spine in ['top', 'right', 'left', 'bottom']:
    ax.spines[spine].set_visible(False)

# 7. Clean Legend positioning
ax.legend(title="Year", frameon=False, loc="upper right")

plt.tight_layout();

```

### Question 16. Create a bar chart showing top 10 products by total revenue

```python
revenue_by_product_name = df.groupby("product_name")["net_revenue"].sum().sort_values(ascending=False).head(10)

ax = (revenue_by_product_name/1e9).plot(kind="bar", figsize=(10,5), title="Top 10 products by total revenue")

ax.bar_label(ax.containers[0], padding=3, fmt='%.1fB')

ax.yaxis.set_visible(False)
ax.margins(y=0.15)
plt.xlabel("Product")
plt.xticks(rotation=45)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)

plt.tight_layout()
;
```

### Question 17. Visualize return rates by product category using a horizontal bar chart

```python
#total sold per category
total_sold_per_category = df.groupby("category")["order_id"].count()

total_sold_per_category
```

```python
#total returned per category
total_returned_per_category = df[df["returned"] == "Yes"].groupby("category")["order_id"].count()

total_returned_per_category
```

```python
# return rate per category
return_rate_per_category = ((total_returned_per_category/total_sold_per_category).fillna(0)*100).sort_values(ascending = False)

return_rate_per_category
```

```python
 
ax = (return_rate_per_category).plot(kind="bar", figsize=(10,5), title="Return rate per category")

ax.bar_label(ax.containers[0], padding=3, fmt='%.1f%%')

ax.yaxis.set_visible(False)
ax.margins(y=0.15)
plt.xlabel("Product")
plt.xticks(rotation=45)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)

# plt.tight_layout()
;
```

### Question 18. Analyse payment method distribution using a pie/donut chart

### Question 20. Build a heatmap showing average revenue by region and category

```python
heatmap_data = pd.pivot_table(
    df,
    values = "gross_revenue",
    index = "region",
    columns = "category",
    aggfunc = "mean"
)

plt.imshow(heatmap_data, cmap="Blues", aspect="auto")
plt.colorbar(label="Average Revenue")
plt.title("Average Revenue by Region and Category")

plt.xticks(range(len(heatmap_data.columns)), heatmap_data.columns, rotation=45)
plt.yticks(range(len(heatmap_data.index)), heatmap_data.index)

plt.tight_layout()

;
```
